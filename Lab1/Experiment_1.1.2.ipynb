```python
# 1. 利用 Tensor 创建两个大小分别 3×2 和 4×2 的随机数矩阵 P 和 Q ，要求服从均值为0，标准差0.01为的正态分布

import torch

# 设置随机种子，以便结果重现
torch.manual_seed(25)

P = torch.normal(mean=0.0, std=0.01, size=(3, 2))
Q = torch.normal(mean=0.0, std=0.01, size=(4, 2))

print("Matrix P (3×2):")
print(P)
print("Shape of P:", P.shape)
print("Mean of P: {:.6f}".format(P.mean().item()))
print("Standard deviation of P: {:.6f}".format(P.std().item()))

print("\nMatrix Q (4×2):")
print(Q)
print("Shape of Q:", Q.shape)
print("Mean of Q: {:.6f}".format(Q.mean().item()))
print("Standard deviation of Q: {:.6f}".format(Q.std().item()))
```

```python
# 2. 对第二步得到的矩阵 Q (4×2) 进行形状变换得到 Q 的转置 Q^T
Q_T = Q.T
# Q_T = Q.transpose(0, 1)
# Q_T = Q.permute(1, 0)

print("Transpose of Q (size: 2×4):")
print(Q_T)
print("Shape of Q^T:", Q_T.shape)
```

```python
# 3. 对上述得到的矩阵 P 和矩阵 Q^T 求矩阵相乘
Res = torch.matmul(P, Q_T)
# Res = P @ Q_T

print("Result R = P × Q^T (3×4):")
print(Res)
print("Shape of Res:", Res.shape)

print(f"Shape of P: {P.shape}")
print(f"Shape of Q^T: {Q_T.shape}") 
```
