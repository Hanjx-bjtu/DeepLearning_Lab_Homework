```python
import torch

# 初始化 x = 1，开启梯度追踪
x = torch.tensor(1.0, requires_grad=True)
print(f"Initial x = {x.item()}")
```

```python
# 计算 y1 = x^2 (保持梯度追踪)
y1 = x ** 2
print(f"y1 = x^2 = {y1.item()}")
```

```python
# 计算 y2 = x^3，但在计算过程中中断梯度追踪
with torch.no_grad():
    y2 = x ** 3
print(f"y2 = x^3 = {y2.item()}")
```

```python
# 计算 y3 = y1 + y2
y3 = y1 + y2
print(f"y3 = y1 + y2 = {y3.item()}")
```

```python
# 计算梯度
y3.backward()
```

```python
# 输出梯度结果
print(f"梯度结果:")
print(f"dy_3/dx = {x.grad.item()}")
print(f"理论值 (如果全程追踪): 2x + 3x^2 = 2×1 + 3×1^2 = 5")
```
